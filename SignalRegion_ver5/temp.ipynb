```python
import os
import pandas as pd

CHANNEL = "1E2Mu"
df = pd.read_csv(f"Outputs/{CHANNEL}/CSV/hyper_params.csv", index_col="mass_point")
```

```python
df
```

```python
for mass_point in df.index:
    path_to = f"Outputs/{CHANNEL}/{mass_point}/best_estimator" 
    #os.makedirs(path_to)
    lr, n_hidden = df.loc[mass_point, "lr_fake"], df.loc[mass_point, "n_hidden_fake"]
    roc_fake = f"Outputs/{CHANNEL}/{mass_point}/figures/roc_sig_vs_fake_learning_rate_{lr}_hidden_layers_{n_hidden}.png"
    training_fake = f"Outputs/{CHANNEL}/{mass_point}/figures/training_sig_vs_fake_learning_rate_{lr}_hidden_layers_{n_hidden}.png"
    os.system(f"mv {roc_fake} {path_to}")
    os.system(f"mv {training_fake} {path_to}")
    
    lr, n_hidden = df.loc[mass_point, "lr_ttX"], df.loc[mass_point, "n_hidden_ttX"]
    roc_ttX = f"Outputs/{CHANNEL}/{mass_point}/figures/roc_sig_vs_ttX_learning_rate_{lr}_hidden_layers_{n_hidden}.png"
    training_ttX = f"Outputs/{CHANNEL}/{mass_point}/figures/training_sig_vs_ttX_learning_rate_{lr}_hidden_layers_{n_hidden}.png"
    os.system(f"mv {roc_ttX} {path_to}")
    os.system(f"mv {training_ttX} {path_to}") 
```
